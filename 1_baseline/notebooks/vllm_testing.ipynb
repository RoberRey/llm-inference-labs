```python
import os

import torch
from huggingface_hub import login
from transformers import AutoTokenizer
from vllm import LLM, SamplingParams
```

```python
login(token=os.getenv('HUGGING_FACE_TOKEN'))
```

```python
torch.cuda.is_available()
```

```python
llm = LLM(model='Qwen/Qwen3-0.6B', gpu_memory_utilization=0.8, max_model_len=10240)
```

```python
tokenizer = AutoTokenizer.from_pretrained('Qwen/Qwen3-0.6B')

prompts = [
    'Hello, my name is',
    'The president of the United States is',
    'The capital of France is',
    'The future of AI is',
]

preprocessed_prompts = []
for prompt in prompts:
    messages = [
        {
            'role': 'system',
            'content': 'You are an useful assistant. ',
        },
        {'role': 'user', 'content': f'{prompt}'},
    ]
    preprocessed_prompt = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
        enable_thinking=False,
    )
    preprocessed_prompts.append(preprocessed_prompt)
```

```python
sampling_params = SamplingParams(temperature=0.8, top_p=0.95, max_tokens=256)
outputs = llm.generate(preprocessed_prompts, sampling_params)

for output in outputs:
    prompt = output.prompt
    generated_text = output.outputs[0].text
    print(f'Prompt: {prompt!r}, Generated text: {generated_text!r}')
```
